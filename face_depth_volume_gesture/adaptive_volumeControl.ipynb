```python
import cv2
import cvzone
import mediapipe
from cvzone.FaceMeshModule import FaceMeshDetector
```

```python
import time
import numpy as np
import osascript
```

```python
cap=cv2.VideoCapture(0)
detect=FaceMeshDetector(maxFaces=1)

while True:
    success,img=cap.read()
    img,faces= detect.findFaceMesh(img,draw=False)
    pTime=0
    cTime = time. time ()
    fps = 1/(cTime-pTime)
    pTime = cTime
    cv2.putText(img,f'FPS: {int(fps)}',(40,70), cv2.FONT_HERSHEY_COMPLEX,1,(255, 0,0),3)
    if faces:
        face = faces[0]
        pointLeft = face[145]
        pointRight = face[374]
        #cv2. line(img, pointLeft, pointRight, (0, 200, 0), 3) 
        #cv2.circle(img, pointLeft, 5, (255, 0, 255), cv2.FILLED) 
        #cv2.circle(img, pointRight, 5, (255, 0, 255), cv2.FILLED)
        w,_ = detect.findDistance(pointLeft,pointRight)
        W=6.3

        #to find focal lenght
        #d=50
        #f=((w*d)/d) 
        #print (f)

        #finding distance
        focal_lenght=840
        d = int((W*focal_lenght) /w)
        print(d)
        
        cvzone.putTextRect(img,f'Depth: {d}cm',(face[10][0]-75,face[10][1]-50))

        minVol=10
        maxVol=100
        lenght=d
        
        vol = np.interp(lenght,[10,100],[minVol, maxVol])

        target_volume = vol
        osascript.osascript("set volume output volume {}".format(target_volume))

    
    cv2.imshow("Image",img)
    cv2.waitKey(1)
```
